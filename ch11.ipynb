```python
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
```

# 11.3

```python
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [67.0, 80.0, 257.0, 1686.0, 6505, 11865.3, 22105.3]

plt.plot(years, gdp, color='green', marker='o', linestyle='solid')
plt.title('GDP per captita')

plt.ylabel('dollars')
plt.savefig('gdp_per_capita.png', dpi=600)
plt.show()
```

# 도전문제 1

```python
array = np.random.randn(1000)
idx = [i for i in range(1000)]

plt.figure(figsize=(40, 5))
plt.plot(idx, array, color='blue', marker='o', linestyle='solid')
plt.title('numbers')

plt.savefig('y=2x.png', dpi=600)
plt.show()
```

# Lab 1

```python
x = [i for i in range(-10, 10)]
y = [i * 2 for i in range(-10, 10)]

plt.plot(x, y, color='blue', marker='o', linestyle='solid')
plt.axis([-20, 20, -20, 20])

plt.ylabel('y')
plt.savefig('y=2x.png', dpi=600)
plt.show()
```

# Lab 2

```python
import math

x = []
y = []

for angle in range(360):
    x.append(angle)
    y.append(math.sin(math.radians(angle)))

plt.plot(x,y)
plt.title("SINE WAVE")
plt.show()
```
